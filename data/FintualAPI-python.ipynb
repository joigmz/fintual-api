# /conceptual_assets/{conceptual_asset_id}/real_assets

```python
headers = {
    'accept': 'application/json',
}
```

```python
df_real_assets = pd.DataFrame()

for i in range(np.shape(df_conceptual_assets.id)[0]):
    c_asset_id = int(df_conceptual_assets.id.iloc[i,])

    response = requests.get('https://fintual.cl/api/conceptual_assets/'+str(c_asset_id)+'/real_assets', headers=headers)

    rt = response.text
    data = json.loads(rt)

    identification,tipo,name,symbol,serie,start_date,end_date,previous_asset_id =[],[],[],[],[],[],[],[]
    conceptual_asset_id = []

    for data in data['data']:
        identification.append(data['id'])
        tipo.append(data['type'])
        name.append(data['attributes']['name'])
        symbol.append(data['attributes']['symbol'])
        serie.append(data['attributes']['serie'])
        start_date.append(data['attributes']['start_date'])
        end_date.append(data['attributes']['end_date'])
        previous_asset_id.append(data['attributes']['previous_asset_id'])
        conceptual_asset_id.append(data['attributes']['conceptual_asset_id'])

    table = np.array([identification,tipo,name,symbol,serie,start_date,end_date,previous_asset_id,conceptual_asset_id]).T

    df_real_assets_i = pd.DataFrame(table, columns=['id','type','name','symbol','serie','start_date','end_date','previous_asset_id','conceptual_asset_id'])

    df_real_assets = pd.concat([df_real_assets,df_real_assets_i], ignore_index=True)

```

```python
df_real_assets.to_csv('df_real_assets.csv',index=False)
df_real_assets
```

# /real_assets/{real_asset_id}/days

* Time interval

```python
real_asset =df_real_assets.iloc[8706,]
real_asset
```

```python
headers = {
    'accept': 'application/json',
}

params = (
    ('from_date', '2013-11-07'),
    ('to_date', '2014-11-10'),
)

real_asset =df_real_assets.iloc[8706,]
real_assets_id = int(real_asset.id)

response = requests.get('https://fintual.cl/api/real_assets/'+str(real_assets_id)+'/days', headers=headers, params=params)

rt = response.text
data = json.loads(rt)

df_real_assets_days = pd.DataFrame()

for i in range(np.shape(data['data'])[0]):
    uno = pd.DataFrame([data['data'][i]['attributes']])
    dos = pd.DataFrame([data['data'][i]['id'],data['data'][1]['type']], index = ['id','type']).T

    day = pd.concat([dos,uno], axis=1)
    
    df_real_assets_days = pd.concat([df_real_assets_days,day], ignore_index=True)

```

```python
df_real_assets_days
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.plot( df_real_assets_days['price'])

name_asset = df_real_assets.loc[df_real_assets['id']==str(real_assets_id)]
plt.title(str(name_asset['name']))

plt.ylabel('price')
plt.xlabel('day')
plt.show()
```
